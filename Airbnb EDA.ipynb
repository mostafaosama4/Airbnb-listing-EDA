```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


data = pd.read_csv('C:\\Users\\hp\\Downloads\\listings.csv')  


summary_stats = data.describe()




plt.figure(figsize=(10, 6))
sns.histplot(data['price'], bins=30, kde=True)
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(data=data, x='room_type', y='availability_365')
plt.title('room types and Availability')
plt.xlabel('room_type')
plt.ylabel('Availability')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(data=data, x='room_type', y='number_of_reviews')
plt.title('room types and number_of_reviews')
plt.xlabel('room_type')
plt.ylabel('number_of_reviews')
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(10, 6))
sns.histplot(data['number_of_reviews'], bins=30, kde=True)
plt.title('Distribution of Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='number_of_reviews', y='price')
plt.title('Comparison between number_of_reviews and price')
plt.xlabel('number_of_reviews')
plt.ylabel('price')
plt.show()

# Correlation Heatmap
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()



```
